```python
import time
import csv
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
```

"Beklemede" Category

```python
driver= webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://www.und.org.tr/afet-destek")
driver.find_element(By.CSS_SELECTOR,"#outermostContainer > div.page-container > div > div > div > div > div.column-wrapper > div.column-1 > div > div.button-wrapper > a").click()
driver.find_element(By.CSS_SELECTOR,  "#outermostContainer > div.page-container > div > div > div.x22-v1-generic-r65-wrapper > div > div.outer-wrapper > div.button-wrapper > div.button-right-wrapper > div.phase-wrapper > div:nth-child(1) > a").click()
total_pages = driver.find_element(By.CSS_SELECTOR, "#outermostContainer > div.page-container > div > div > div.g5-v1-metrostyle-r5-wrapper > div > div > span:nth-child(9) > a").text
current_url = driver.current_url
print(current_url)
```

```python
base_url = 'https://www.und.org.tr/afet-destek/yukum_var'
#first page
page_number = 0
url = f'{base_url}?page={page_number}&phaseID=18692' #&phase... term depends on the category of shipment
driver.get(url)
total_pages = driver.find_element(By.CSS_SELECTOR, "#outermostContainer > div.page-container > div > div > div.g5-v1-metrostyle-r5-wrapper > div > div > span:nth-child(9) > a").text
#empty list to store all URL's
beklemede_buttons_links_all = []
# paginate
while page_number < int(total_pages):
    try:
        page_number += 1
        url = f'{base_url}?page={page_number}&phaseID=18692'
        driver.get(url)
        beklemede_bottons_list = driver.find_elements(By.CLASS_NAME,"detail-button-wrapper")
        for b in beklemede_bottons_list:
            beklemede_buttons_links_all.append(b.find_element(By.TAG_NAME, "a").get_attribute("href"))
        print(f"Page {page_number}")
        time.sleep(10)
    except:
        break
data = list(zip(beklemede_buttons_links_all)) 
with open("D:\deprem veri\youtube_data\shipments-deprem-urls-latest.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Beklemede"])
    for row in data:
        writer.writerow(row)

driver.quit()
```

```python
no = []
tarih= []
mal_cinsi =[]
miktar = []
yuklenecek_yer = []
gidecegi_yer = []
gonderici_kurum = []
kurum_yetkili = []
tel = []
email = []
aciklama = []
urls =[]
mesaj = []
mesaj_tarihi  = []
for url in beklemede_buttons_links_all:
    driver= webdriver.Chrome(service=Service(ChromeDriverManager().install()))
    driver.get(url)
   
   
    try:
        no_val = driver.find_element(By.CSS_SELECTOR, "#outermostContainer > div.page-container > div > div > div > div > div.form-detail-wrapper > div > div:nth-child(1) > span:nth-child(3)")
        no.append(no_val.text)
    except:
        no.append("NA")


    
    try:  
        tarih_val = driver.find_element(By.CSS_SELECTOR, "#outermostContainer > div.page-container > div > div > div > div > div.form-detail-wrapper > div > div:nth-child(2) > span:nth-child(3)")
        tarih.append(tarih_val.text)
    except:
        tarih.append("NA")



    try:   
        mal_val = driver.find_element(By.CSS_SELECTOR, "#outermostContainer > div.page-container > div > div > div > div > div.form-detail-wrapper > div > div:nth-child(3) > span:nth-child(3)")
        mal_cinsi.append(mal_val.text)
    except:
        mal_cinsi.append("NA")


    try:
        miktar_val = driver.find_element(By.CSS_SELECTOR, "#outermostContainer > div.page-container > div > div > div > div > div.form-detail-wrapper > div > div:nth-child(4) > span:nth-child(3)")
        miktar.append(miktar_val.text)
    except:
        miktar.append("NA")
        

    try: 
        yuklenecek_val = driver.find_element(By.CSS_SELECTOR, "#outermostContainer > div.page-container > div > div > div > div > div.form-detail-wrapper > div > div:nth-child(5) > span:nth-child(3)")
        yuklenecek_yer.append(yuklenecek_val.text)
    except:
        yuklenecek_yer.append("NA")

    try: 

        gidecegi_val = driver.find_element(By.CSS_SELECTOR, "#outermostContainer > div.page-container > div > div > div > div > div.form-detail-wrapper > div > div:nth-child(6) > span:nth-child(3)")
        gidecegi_yer.append(gidecegi_val.text)
    except:
        gidecegi_yer.append("NA")


    try:  
        gonderici_val = driver.find_element(By.CSS_SELECTOR, "#outermostContainer > div.page-container > div > div > div > div > div.form-detail-wrapper > div > div:nth-child(7) > span:nth-child(3)")
        gonderici_kurum.append(gonderici_val.text)
    except:
        gonderici_kurum.append("NA")

    try: 
        kurum_val = driver.find_element(By.CSS_SELECTOR, "#outermostContainer > div.page-container > div > div > div > div > div.form-detail-wrapper > div > div:nth-child(8) > span:nth-child(3)")
        kurum_yetkili.append(kurum_val.text)
    except: 
        kurum_yetkili.append("NA")

    try:
        tel_val = driver.find_element(By.CSS_SELECTOR, "#outermostContainer > div.page-container > div > div > div > div > div.form-detail-wrapper > div > div:nth-child(9) > span:nth-child(3)")
        tel.append(tel_val.text)
    except:
        tel.append("NA")

    try:  
        email_val = driver.find_element(By.CSS_SELECTOR, "#outermostContainer > div.page-container > div > div > div > div > div.form-detail-wrapper > div > div:nth-child(10) > span:nth-child(3)")
        email.append(email_val.text)
    except:
        email.append("NA")

    try: 
        aciklama_val = driver.find_element(By.XPATH, '//*[@id="outermostContainer"]/div[1]/div/div/div/div/div[3]/div/div[10]/span[3]/text()')
        aciklama.append(aciklama_val.text)
    except:
        aciklama.append("NA")

    try:

        mesaj_val = driver.find_element(By.CSS_SELECTOR, "#outermostContainer > div.page-container > div > div > div > div > div.connected-forums-wrapper > div.connected-forums-inner > div > div:nth-child(2)")
        mesaj.append(mesaj_val.text)
    except:
        mesaj.append("NA")
        
    try:

        mesaj_tarih_val = driver.find_element(By.CSS_SELECTOR, "#outermostContainer > div.page-container > div > div > div > div > div.connected-forums-wrapper > div.connected-forums-inner > div > div:nth-child(1) > div.connected-forum-col-date")   
        mesaj_tarihi.append(mesaj_tarih_val.text)
    except:
        mesaj_tarihi.append("NA")
        
        
    urls.append(url)  
    time.sleep(10)

data = list(zip(urls,no,tarih,mal_cinsi,miktar,yuklenecek_yer,gidecegi_yer,gonderici_kurum,kurum_yetkili,tel,email,aciklama,mesaj,mesaj_tarihi))

with open("D:\deprem veri\youtube_data\dave-gahan.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["url", "form_id","tarih","mal_cinsi","miktar","yuklenecek_yer","gidecegi_yer","gonderici_kurum","kurum_yetkili","tel","email","aciklama","mesaj","mesaj_tarihi"])
    for row in data:
        writer.writerow(row)
```

# Folium Visualisation

```python
#geocoding
from geopy.geocoders import Nominatim
geolocator = Nominatim()
longitude = []
latitude = []

for city in info_data["yuklenecek_yer"]:
    loc = geolocator.geocode(city)
    if loc is None:
        latitude.append("NAN")
        longitude.append("NAN")
    else:
        latitude.append(loc.latitude)
        longitude.append(loc.longitude)

info_data["Longitude"] = longitude
info_data["Latitude"] = latitude

print(info_data)
```

```python
info_data.dropna(axis=0, subset=['Longitude', 'Latitude']) #dropping NA values for visualisation
```

```python
import folium
# create map
map_center = [info_data["Latitude"].iloc[0], info_data["Longitude"].iloc[0]]
m = folium.Map(location=map_center, zoom_start=6)

for i, row in info_data.iterrows(): #markers
    popup_text = "<b>Yük Bilgisi:</b><br>"
    for col in info_data.columns:
        popup_text += f"{col}: {row[col]}<br>"   
    popup = folium.Popup(popup_text, max_width=300)
    folium.Marker([row["Latitude"], row["Longitude"]], popup=popup).add_to(m)
m
```
